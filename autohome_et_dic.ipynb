```python
import requests
from requests import Request,Session
from bs4 import BeautifulSoup
import collections
from itertools import chain
import pandas as pd
from collections import Counter
```

```python
dic_et = {1:'汽油',
          2:'柴油',
          3:'油电混合',
          4:'纯电动',
          5:'插电式混动'}
```

```python
def get_results(dic_et):
    result_dic = {}
    for k in dic_et.keys():
        url = f'https://www.autohome.com.cn/car/0_0-0.0_0.0-0-0-0-0-{k}-0-0-0/'
        get_model_id_dic(url,dic_et[k],result_dic) # 动态更新result_dic
    return result_dic
```

```python
def get_model_id_dic(url,energytype,result_dic):
    
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) ' \
         'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}
    result = requests.get(url,headers=headers)
    bsobj = BeautifulSoup(result.content,'lxml')
    uls_model = bsobj.find_all('ul',{'class':'rank-list-ul'})
    
    for ul_model in uls_model:
        lis_model = ul_model.find_all('li')
        for li_model in lis_model:
            if li_model.has_attr('id'):
                model_id = li_model.attrs['id'][1:]
                
                if not model_id in result_dic:
                    result_dic[model_id]={'汽油':None,
                                          '柴油':None,
                                          '油电混合':None,
                                          '纯电动':None,
                                          '插电式混动':None}
                result_dic[model_id][energytype] = True
                
```

```python
result_dic = get_results(dic_et)
```
